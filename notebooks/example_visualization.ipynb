# Visualización y generación de ejemplo
Tras entrenar los modelos, utilizaremos las redes de "Decoder" de ambas arquitecturas para generar números. Definiremos aleatoriamente el vector latente.

```python
import matplotlib.pyplot as plt
import torch
import os
import sys
from src import model

conv_model = model.ConvCVAE.load_from_checkpoint("../models/main/best_model-1.0.ckpt")
conv_model.eval()

fig, axes = plt.subplots(10, 10, figsize=(10, 10)) 

with torch.no_grad():
    for num_to_gen in range(10):
        for i in range(10): 
            sample = conv_model.generate_number(num_to_gen)
            axes[num_to_gen, i].imshow(sample, cmap='gray')
            axes[num_to_gen, i].axis('off')
            
            if i == 0:
                axes[num_to_gen, 0].set_ylabel(f"Dígito {num_to_gen}", rotation=0, labelpad=40, fontsize=12, fontweight='bold')

fig.suptitle('Imágenes Generadas por el modelo', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0.05, 0, 1, 0.95])
plt.show()
```
